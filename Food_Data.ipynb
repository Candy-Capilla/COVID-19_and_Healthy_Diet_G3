```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/Food_Supply_kcal_Data.csv"
food_supply_df = pd.read_csv(csv_file)
food_df= food_supply_df[food_supply_df.columns[:-8]].set_index('Country')
food_df
```

```python
covid_df= food_supply_df[["Country","Confirmed", "Deaths","Recovered", "Active", "Population"]].set_index("Country")
covid_df
```

```python
grains= food_supply_df[["Cereals - Excluding Beer", "Oilcrops", "Spices"]].sum(axis=1)
grains
```

```python
vegs=food_supply_df[["Pulses", "Starchy Roots", "Sugar Crops", "Vegetal Products", "Vegetables"]].sum(axis=1)
vegs
```

```python
fruits=food_supply_df[["Fruits - Excluding Wine", "Treenuts"]].sum(axis=1)
fruits
```

```python
protein=food_supply_df[["Animal Products", "Animal fats", "Aquatic Products, Other", "Meat", "Milk - Excluding Butter", "Offals", "Eggs", "Fish, Seafood"]].sum(axis=1)
protein
```

```python
misc=food_supply_df[["Miscellaneous", "Stimulants", "Sugar & Sweeteners", "Vegetable Oils", "Alcoholic Beverages"]].sum(axis=1)
misc
```

```python
clean_food=pd.DataFrame({'Grains': grains, 'Vegetables':vegs, 'Fruits':fruits, 'Proteins':protein, "Misc":misc})
clean_food
```

```python
country= food_supply_df[['Country']]
clean_food_df= country.join(clean_food).set_index("Country")
clean_food_df
```

```python
clean_food_df["Total"]= clean_food_df.sum(axis=1)
clean_food_df
```

```python
#double check numbers
food_df.sum(axis=1)
```

```python
#Merge food and covid df
food_covid_df= pd.merge(clean_food_df, covid_df, on= "Country", how= "inner")
food_covid_df
```
