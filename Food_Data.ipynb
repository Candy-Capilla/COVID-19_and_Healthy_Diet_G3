```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "Resources/Food_Supply_kcal_Data.csv"
food_supply_df = pd.read_csv(csv_file)
food_df= food_supply_df[food_supply_df.columns[:-8]]
food_df
```

```python
food_supply_df= food_supply_df.copy()
food_supply_df.rename(columns={"Country":"country"}, inplace= True)
food_supply_df
```

```python
covid_df= food_supply_df[["country","Confirmed", "Deaths","Recovered", "Active", "Population"]].set_index("country")
covid_df
```

```python
grains= food_supply_df[["Cereals - Excluding Beer", "Oilcrops", "Spices"]].sum(axis=1)
grains
```

```python
vegs=food_supply_df[["Pulses", "Starchy Roots", "Sugar Crops", "Vegetal Products", "Vegetables"]].sum(axis=1)
vegs
```

```python
fruits=food_supply_df[["Fruits - Excluding Wine", "Treenuts"]].sum(axis=1)
fruits
```

```python
protein=food_supply_df[["Animal Products", "Animal fats", "Aquatic Products, Other", "Meat", "Milk - Excluding Butter", "Offals", "Eggs", "Fish, Seafood"]].sum(axis=1)
protein
```

```python
misc=food_supply_df[["Miscellaneous", "Stimulants", "Sugar & Sweeteners", "Vegetable Oils", "Alcoholic Beverages"]].sum(axis=1)
misc
```

```python
clean_food=pd.DataFrame({'Grains': grains, 'Vegetables':vegs, 'Fruits':fruits, 'Proteins':protein, "Misc":misc})
clean_food
```

```python
country= food_supply_df[['country']]
clean_food_df= country.join(clean_food).set_index("country")
clean_food_df
```

```python
clean_food_df["Total"]= clean_food_df.sum(axis=1)
clean_food_df
```

```python
#double check numbers
food_df.sum(axis=1)
```

```python
#Merged df
food_covid_df= pd.merge(clean_food_df, covid_df, on= "country", how= "inner")
food_covid= food_covid_df.rename(columns={"Grains":"grains", "Vegetables":"vegetables", "Fruits":"fruits", "Proteins":"proteins", "Misc":"misc", "Total":"total", "Confirmed":"confirmed", "Deaths": "deaths", "Recovered": "recovered", "Active":"active", "Population": "population"})
food_covid
```

```python
#Call next source
csv_file = "Resources/Food_Supply_Quantity_kg_Data .csv"
quantity_food_df = pd.read_csv(csv_file)
quantity_df= quantity_food_df
quantity_df
```

```python
quantity_df= quantity_df.copy()
quantity_df.rename(columns={"Country":"country"}, inplace= True)
quantity_df
```

```python
clean_quantity_df= quantity_df[["country","Obesity", "Undernourished", "Confirmed", "Deaths", "Recovered", "Active", "Population"]].set_index("country")
clean_quantity_df
```

```python
lowercase_quantity_df= clean_quantity_df.rename(columns={"Obesity":"obesity","Undernourished":"undernourished", "Confirmed":"confirmed", "Deaths": "deaths", "Recovered":"recovered", "Active": "active", "Population": "population"})
lowercase_quantity_df
```

```python
connection_string = "postgres:postgres@localhost:5432/food_covid_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#check for tables
engine.table_names()
```

```python
food_covid.to_sql(name= 'food_covid', con=engine, if_exists= 'append', index=True)
```

```python
lowercase_quantity_df.to_sql(name= 'quantity', con=engine, if_exists= 'append', index=True)
```
